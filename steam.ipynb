```python
english=[0, 1]
df['english'] = pd.Categorical(df['english'], categories=english, ordered=False)

df['release_date'] = pd.to_datetime(df['release_date'])
df['releaste_year'] = df['release_date'].dt.year

df['release_month'] = df['release_date'].dt.month
month=range(1,13)
df['release_month'] = pd.Categorical(df['release_month'], categories=month, ordered=True)

df['release_day'] = df['release_date'].dt.day
day=range(1,31)
df['release_day'] = pd.Categorical(df['release_day'], categories=day, ordered=True)

df['developer'] = df['developer'].astype('category')
df['publisher'] = df['publisher'].astype('category')
```
