**3. Add a column that holds the total screen resolution for each device. Name it resolution.**


```python
resolution = mobiles["px_height"] * mobiles["px_width"]
mobiles["resolution"] = resolution
mobiles.head()
```

a function that replaces inf / -inf values with NaN (incase of zero divison)

```python
def infToNan(x_df, colmn_name):
  """ replace inf or -inf values in a certain column to NaN """
  x_df.loc[~np.isfinite(mobiles[colmn_name]), colmn_name] = np.nan
```

**4. Add a column that holds the DPI (dots per inch) of the screen width and name it DPI_w.**

```python
DPI_w = mobiles['px_width']/(mobiles['sc_w']/2.54)
mobiles["DPI_w"] = DPI_w
infToNan(mobiles, 'DPI_w')
mobiles.head()
```

**5. Add a column that holds the ratio battery_power/talk_time and name it call_ratio.**

```python
call_ratio = mobiles["battery_power"].div(mobiles["talk_time"])
mobiles["call_ratio"] = call_ratio
infToNan(mobiles, 'call_ratio')
mobiles.head()
```

**6. Change the memory column to hold the memory in GB instead of MB.**

**8. Convert the following features into categorical series in the Dataframe: speed,screen,cores**

```python
speeds = ['low', 'medium', 'high']
mobiles['speed'] = pd.Categorical(mobiles['speed'], categories=speeds, ordered=True)

screens = ['Touch', 'LCD']
mobiles['screen'] = pd.Categorical(mobiles['screen'], categories=screens, ordered=False)

cores=['single', 'dual', 'triple', 'quad', 'penta', 'hexa', 'hepta', 'octa']
mobiles['cores'] = pd.Categorical(mobiles['cores'], categories=cores, ordered=True)

mobiles.dtypes
```

**1. How many phones do not have a camera at all (front or back)?**

**2. What is the average battery power for single-sim phones that have a camera or front camera
with a higher resolution than 12 megapixels?**

```python
average_power = mobiles[(mobiles['sim'] == 'Single') & (mobiles['camera'].notna() | (mobiles['f_camera']>12))]
average_power["battery_power"].mean()
```

**3. What is the ID and price of the most expensive phone that has no wifi, a touch screen and
weighs more than 145 grams?**

```python
exp_price = mobiles[(mobiles['wifi'] == 'none') & (mobiles['screen']=='Touch') & (mobiles['mobile_wt']>145)]
exp_price = exp_price.sort_values(['price'], ascending=False)
exp_price[['price']].head(1)
```

**4. Create a pivot table that shows the percentage of phones with Bluetooth per generation,
pivoted around the phone generation and split by “ram” quartiles. (i.e. the rows are the
generation number and the columns are 4 quartiles of ram size).**

```python
mobiles['bluetooth01'] = mobiles['bluetooth'].map({'Yes': 1, 'No': 0})
ram = pd.qcut(mobiles['ram'], 4)
mobiles.pivot_table('bluetooth01', ['gen'], [ram])
```

**5. Create a new Dataframe based on the original that has the following features: [id,
battery_power, ram, talk_time, Bluetooth, cores, sim, memory, price], and contains a random
sampling of half of the medium speed phones**

```python
df = mobiles[mobiles['speed'] == 'medium'].sample(frac = 0.5).copy()
df.insert(0, 'id', df.index)
df.reset_index(drop=True, inplace=True)
df.index.name = '#'
df = df[['id', 'battery_power', 'ram', 'talk_time', 'bluetooth', 'cores', 'sim', 'memory', 'price']]
df
```

**6. Using this new dataset, what is the maximum total talk time you can achieve if you use 3
